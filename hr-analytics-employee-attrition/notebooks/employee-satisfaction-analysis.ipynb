### HR Analytics Employee Attrition - Data Visualization

```python
import pandas as pd #import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

```python
data = pd.read_csv('IBM-HR-Employee-Attrition.csv') #load data
data.head(5) #display first 5 rows
```

```python
data.shape #number of observations (1470 samples, 35 columns)
```

```python
data.columns #columns name
```

```python
data.describe() #describe numeric data fields
```

```python
#check for null values
100*data.isnull().sum()/data.shape[0] #data has no missing values
```

```python
#Number of active and inactive employee by each department
pd.pivot_table(data, index = 'Attrition', values = 'EmployeeCount', columns = ['Department'],aggfunc = 'count')
```

```python
#Descriptive statistics of active and inactive employees:
    #Active employees had a higher average monthly income
    #Active employees had slightly higher training times last year and better work life balance
pd.pivot_table(data, index = 'Attrition',values = ['MonthlyIncome','TrainingTimesLastYear','WorkLifeBalance'],aggfunc='mean')
```

```python
#drop some unmeaningful, duplicate data
data.drop(['EmployeeNumber','StandardHours','HourlyRate','EmployeeCount'],axis=1,inplace=True)
data.head()
```

```python
# Find correlations for Employee attrition and sort
# Individually, employee attrition has weak correlation with other variables
df_HR = data.copy()
df_HR['Target'] = df_HR['Attrition'].apply(lambda x: 0 if x == 'No' else 1)
df_HR = df_HR.drop(['Attrition'], axis=1)
correlations = df_HR.corr()['Target'].sort_values()
print('Most Positive Correlations: \n', correlations.tail(5))
print('\nMost Negative Correlations: \n', correlations.head(5))
```

```python
# Correlations visualization
corr = df_HR.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
# Heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr,vmax=.5,
            mask=mask, annot=True, fmt='.2f',
            linewidths=.2, cmap="YlGnBu")
plt.show()
```

```python
data.hist(figsize=(20,20)) #frequency distribution for numeric data fields
plt.show()
```

```python
#Younger people quit more
plt.figure(figsize=(16,3))
sns.kdeplot(data.loc[data['Attrition']=='No', 'Age'],label='Active Staff')
sns.kdeplot(data.loc[data['Attrition']=='Yes','Age'],label='Nonactive Staff')
plt.title('Age distribution (%) in Attrition Status')
plt.xlabel('Age')
plt.legend()

plt.xlim(left=0)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(16,9))

#Non-active staffs tended to quit after working at the company for 5 years
sns.kdeplot(data.loc[data['Attrition']=='No', 'YearsAtCompany'],label='Active Staff',ax=ax2)
sns.kdeplot(data.loc[data['Attrition']=='Yes','YearsAtCompany'],label='Nonactive Staff',ax=ax2)
ax2.set_title('Years at company distribution (%) in attrition status')
ax2.set_xlabel('Year')
ax2.legend()

#Active and non-active staffs tended to have the smae promotion period
sns.kdeplot(data.loc[data['Attrition']=='No', 'YearsSinceLastPromotion'],label='Active Staff',ax=ax1)
sns.kdeplot(data.loc[data['Attrition']=='Yes','YearsSinceLastPromotion'],label='Nonactive Staff',ax=ax1)
ax1.set_title('Years since last promotion distribution (%) in attrition status')
ax1.set_xlabel('Year')
ax1.legend()

plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.show()
```

```python
#Male employees had higher monthly income than female employees
sns.displot(data = data, x='MonthlyIncome',col='Attrition',kind = 'kde',hue='Gender')
plt.xlabel('Income')
plt.xlim(left=0)
plt.show()
```

```python
#Leavers by education fields: HR degree witnessed the highest retention rate
x = {}
for i in data['EducationField'].unique():
    re = data[(data['EducationField']==i)&(data['Attrition']=="Yes")].shape[0]/data[data['EducationField']==i].shape[0]
    x[i] = re*100

plt.figure(figsize=(9,5))
plt.bar(*zip(*x.items()))
plt.title('Leavers by Education Fields')
plt.xlabel('Education field')
plt.ylabel('%')

plt.show()
```

```python
#Leavers by gender: Male employees quitted more
df_leaver = pd.DataFrame(columns = ['Gender','% leave'])
n =0
for i in data['Gender'].unique():
    re = data[(data['Gender']==i)&(data['Attrition']=="Yes")].shape[0]/data[data['Gender']==i].shape[0]
    df_leaver.loc[n] = (i,re*100)
    n += 1
    
plt.figure(figsize=(16,4))
df_gen = df_leaver.groupby(by='Gender').sum()
df_gen.plot(kind='bar',title='Leavers by Gender')
plt.xlabel('Gender')
plt.ylabel('%')

plt.show()
```

```python
#leaver by job roles: Sale representative had the highest retention rate
x = {}
for i in data['JobRole'].sort_values(ascending=False).unique():
    re = data[(data['JobRole']==i)&(data['Attrition']=="Yes")].shape[0]/data[data['JobRole']==i].shape[0]
    x[i] = re*100

plt.figure(figsize=(17,5))
plt.bar(*zip(*x.items()))
plt.title('Leavers by Job Role')
plt.xlabel('Job Role')
plt.ylabel('%')
plt.show()
```
